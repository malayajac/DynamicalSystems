Logistic map: $x_{n+1} = \mu x_{n} (1 - x_{n})$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def LogisticFn(mu, x):
    return mu * x * (1 - x)
```

```python
x = np.linspace(0, 1, 1001)
```

```python
mu = 3.4
```

```python
f = LogisticFn(mu, x)
```

#### Return Map

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, f)
ax.plot(x, x, "k")

ax.set_xlabel("$x_{n}$", fontsize=20)
ax.set_ylabel("$x_{n+1}$", fontsize=20)
ax.set_title("$\mu$={}".format(mu), fontsize=20)

ax.grid()
plt.show()
```

#### Bifurcation Diagram

```python
dis = 5000
N = 100
```

```python
%%time
#---
mu_List = np.linspace(0, 4, 1001)
#---
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
#---
for mu in mu_List:
    x_List = [np.random.random()]
    for ii in range(1, dis+N):
        xtemp = LogisticFn(mu, x_List[ii-1])
        x_List.append(xtemp)
    ax.plot([mu]*N, x_List[dis:], "k,")
#---
ax.set_xlim([0, 4])
ax.set_ylim([0, 1])

ax.set_xlabel("$\mu$", fontsize=20)
ax.set_ylabel("x*", fontsize=20)

ax.grid()
plt.show()
```
