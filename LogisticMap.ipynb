Logistic map: $x_{n+1} = \mu x_{n} (1 - x_{n})$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def LogisticFn(mu, x):
    return mu * x * (1 - x)
```

```python
x = np.linspace(0, 1, 1001)
```

```python
mu = 3.4
```

```python
f = LogisticFn(mu, x)
```

#### Return Map

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, f)
ax.plot(x, x, "k")

ax.set_xlabel("$x_{n}$", fontsize=20)
ax.set_ylabel("$x_{n+1}$", fontsize=20)
ax.set_title("$\mu$={}".format(mu), fontsize=20)

ax.grid()
plt.show()
```
