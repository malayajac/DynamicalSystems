The Mandelbrot set is the set of complex numbers $c$ for which the function $f_{c}(z) = z^{2} + c$ does not diverge when iterated from $z = 0$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%%time
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

w_left, w_right = -2.5, 1.0
w_bottom, w_top = -1.5, 1.5

for y in np.linspace(w_bottom, w_top, 101):
    for x in np.linspace(w_left, w_right, 101):
        c = np.complex(x, y)
        z = 0
        for ii in range(100):
            if abs(z)>2:
                ax.plot(x, y, "k.", alpha=0.5)
                break
            z = z*z + c

ax.set_xlim([w_left, w_right])
ax.set_ylim([w_bottom, w_top])
ax.grid(True)
plt.show()
```
