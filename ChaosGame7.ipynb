```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Initialize coods of vertices of a square of side length=1:
v1 = [0.0, 0.0]
v2 = [1.0, 0.0]
v3 = [1.0, 1.0]
v4 = [0.0, 1.0]

# Midpoints of each side:
v12 = [0.5, 0.0]
v23 = [1.0, 0.5]
v34 = [0.5, 1.0]
v41 = [0.0, 0.5]
```

```python
# N = 10**4 #number of points I will plot
# N = 2*10**4 #number of points I will plot
N = 4*10**4 #number of points I will plot
```

```python
# IC:
px = np.random.random()
py = np.random.random()
print(px, py)
```

```python
CoinTossFaces = set([1, 2, 3, 4, 5, 6, 7, 8])
```

```python
%%time
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(v1[0], v1[1], "k,")
ax.plot(v2[0], v2[1], "k,")
ax.plot(v3[0], v3[1], "k,")
ax.plot(v4[0], v4[1], "k,")
ax.plot(v12[0], v12[1], "k,")
ax.plot(v23[0], v23[1], "k,")
ax.plot(v34[0], v34[1], "k,")
ax.plot(v41[0], v41[1], "k,")

for ii in range(N):
    
    toss = np.random.choice(list(CoinTossFaces))
    
    if toss==1:
        pxnew = (1./3)*(px - v1[0]) + v1[0]
        pynew = (1./3)*(py - v1[1]) + v1[1]
    
    elif toss==2:
        pxnew = (2./3)*(v2[0] - px) + px
        pynew = (1./3)*(py - v2[1]) + v2[1]
    
    elif toss==3:
        pxnew = (2./3)*(v3[0] - px) + px
        pynew = (2./3)*(v3[1] - py) + py

    elif toss==4:
        pxnew = (1./3)*(px - v4[0]) + v4[0]
        pynew = (2./3)*(v4[1] - py) + py
    
    elif toss==5:
        pxnew = (2./3)*(v12[0] - px) + px
        pynew = (1./3)*(py - v12[1]) + v12[1]
    
    elif toss==6:
        pxnew = (2./3)*(v23[0] - px) + px
        pynew = (1./3)*(py - v23[1]) + v23[1]

    elif toss==7:
        pxnew = (2./3)*(v34[0] - px) + px
        pynew = (2./3)*(v34[1] - py) + py
    
    elif toss==8:
        pxnew = (1./3)*(px - v41[0]) + v41[0]
        pynew = (1./3)*(py - v41[1]) + v41[1]
    
    px = pxnew
    py = pynew
    
    ax.plot(px, py, "k,")

ax.grid()
ax.set_aspect("equal")
plt.show()
```
